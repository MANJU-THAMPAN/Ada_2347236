```python
import pandas as pd
import matplotlib.pyplot as plt

# Aggregating the data by summing the crime counts across all major and minor heads
trend_data = {
    'Time Period': [
        'Current Year (Upto Month End)',
        'Corresponding Month Previous Year',
        'Previous Month',
        'Current Month'
    ],
    'Total Crimes': [
        df['During the current year upto the end of month under review'].sum(),
        df['During the corresponding month of previous year'].sum(),
        df['During the previous month'].sum(),
        df['During the current month'].sum()
    ]
}

trend_df = pd.DataFrame(trend_data)

# Plotting the trend analysis
plt.figure(figsize=(10, 6))
plt.plot(trend_df['Time Period'], trend_df['Total Crimes'], marker='o', linestyle='-', color='b')
plt.title('Trend Analysis of Crime Data')
plt.xlabel('Time Period')
plt.ylabel('Total Number of Crimes')
plt.grid(True)
plt.show()
```

```python


from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Sl.No.']
  ys = series['During the current year upto the end of month under review']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Sl.No.', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Heads of Crime')):
  _plot_series(series, series_name, i)
  fig.legend(title='Heads of Crime', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Sl.No.')
_ = plt.ylabel('During the current year upto the end of month under review')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error


# Data Cleaning and Preprocessing
df.fillna(0, inplace=True)

```

```python

# Prepare data for prediction
X = df[['During the corresponding month of previous year', 'During the previous month']]
y = df['During the current month']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define models
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(n_estimators=100),
    'SVM': SVR(),
    'Gradient Boosting': GradientBoostingRegressor()
}

# Initialize a dictionary to store the results
results = {}
```

```python
# Train and evaluate each model
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Calculate metrics
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mse)
    mape = mean_absolute_percentage_error(y_test, y_pred)

    # Store results
    results[model_name] = {
        'MSE': mse,
        'MAE': mae,
        'R2 Score': r2,
        'RMSE': rmse,
        'MAPE': mape
    }

# Display the results
results_df = pd.DataFrame(results).T
print(results_df)

# Plotting the results for comparison
results_df.plot(kind='bar', figsize=(14, 8))
plt.title('Model Performance Comparison')
plt.ylabel('Metric Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Initialize a DataFrame to store actual and predicted values
predictions_df = pd.DataFrame(y_test).reset_index(drop=True)
predictions_df.columns = ['Actual']

# Train and evaluate each model
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Add the predicted values to the DataFrame
    predictions_df[model_name] = y_pred

# Display the table of actual vs predicted values
print(predictions_df)

# Save the table as a CSV file if needed
predictions_df.to_csv('actual_vs_predicted.csv', index=False)
```
